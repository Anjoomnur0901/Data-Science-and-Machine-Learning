```python
diabetes_Status = df['diabetes'].value_counts()
transactions = diabetes_Status.index
quantity = diabetes_Status.values

figure = px.pie(df,
             values=quantity,
             names=transactions, hole =.60,
             title="diabetes Status")
figure.show()
```

<h3>The value_counts()</h3><p> The value_counts() method is used on the 'diabetes' column of the DataFrame df. This method counts the number of occurrences of each unique value in the 'diabetes' column and returns a Series with these counts.</p><br>
<h3>diabetes_Status.index</h3><p> diabetes_Status.index retrieves the unique values from the 'diabetes' column, which are the different diabetes statuses.</p><br>
<h3>diabetes_Status.values</h3><p>diabetes_Status.values retrieves the counts of each unique value, representing the quantity of each diabetes status.</p><br>

<p>A pie chart is created using the px.pie function from the Plotly Express module (px).<br>
<b>values=quantity</b> specifies the sizes of the slices of the pie chart, which correspond to the counts of each diabetes status.<br>
<b>names=transactions</b> specifies the labels for the slices, which correspond to the unique diabetes statuses.<br>
<b>hole=.60</b> creates a donut chart by setting the hole size to 60% of the pie chart's radius.</br>
<b>title="diabetes Status"</b> sets the title of the pie chart.</p>

```python
# Get pie chart slice labels
pie_chart_slice_labels = df['diabetes'].value_counts().index.to_numpy()

# Get pie chart slice values
pie_chart_slice_values = df['diabetes'].value_counts().to_numpy()

# Create a pie chart of the number of customers for each country origin
plt.pie(pie_chart_slice_values, labels=pie_chart_slice_labels, autopct="%1.1f%%")
plt.title("Pie Chart of Diabetes")
plt.show()
```

<b>df['diabetes'].value_counts().index.to_numpy()</b> counts the unique values in the 'diabetes' column of the DataFrame df, retrieves the index (unique values), and converts it to a NumPy array. This array will be used as the labels for the pie chart slices.<br>
<b>df['diabetes'].value_counts().to_numpy()</b> counts the unique values in the 'diabetes' column, retrieves the counts, and converts them to a NumPy array. These values will be used as the sizes of the pie chart slices.</br>
<b>plt.pie()</b> is a function from Matplotlib that creates a pie chart.</br>
<b>pie_chart_slice_values</b> specifies the sizes of the pie chart slices.<br>
<b>labels=pie_chart_slice_labels</b> specifies the labels for the slices.<br>
<b>autopct="%1.1f%%"<b> formats the percentage labels on the pie chart slices to display one decimal place.

<h1>Train Test Split</h1>

```python
 x = df.drop('DEATH_EVENT',axis=1)
 y = df[['DEATH_EVENT']]
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, random_state=42, train_size=.70)
```

<b>random_state=42:</b> This sets a seed for the random number generator to ensure reproducibility of the split. Using the same random_state will always produce the same train-test split.<br>
<b>train_size=0.70:</b> This specifies that 70% of the data should be used for training, and the remaining 30% will be used for testing.

```python
xtest.shape
```

```python
ytest.shape
```

```python
ytest.head(3)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(xtrain, ytrain) #training done
```

```python
#let's predict
pred = dt.predict(xtrain) #training result
pred
```

```python
import numpy as np
np.array(ytrain['DEATH_EVENT'])
```

```python
ytrain['Predicted_Event'] = pred
```

```python
ytrain.head(10)
```

```python
dt.score(xtrain, ytrain['DEATH_EVENT'])
```
