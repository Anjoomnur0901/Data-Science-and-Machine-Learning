<h1>Axes labels, Legend, Grid</h1>

```python
days = [1,2,3,4,5,6,7]
max_temperature = [50,51,52,48,47,49,46]
min_temperature = [43,42,40,44,33,35,37]
avg_temperature = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days') 
plt.ylabel('Temperature')
plt.title("Weather")
plt.plot(days,max_temperature, label="Max")
plt.plot(days,min_temperature, label="Min")
plt.plot(days,avg_temperature,label="Avg")

plt.legend(loc="best", shadow=True, fontsize="large")
plt.grid()
```

<h1>Histograms</h1>
<h3>Histograms are bar charts where the y axis contains the frequency of a variable<h3>

```python
#How many patients are normal, prediabetic and diabetic?
blood_sugar_man=[113,85.90,150,149,89,93,115,80,77,82,129]
```

```python
plt.hist(blood_sugar_man)
```

```python
plt.hist(blood_sugar_man, bins = 3)
```

```python
plt.hist(blood_sugar_man, bins = 3, rwidth= 0.95)
```

```python
plt.hist(blood_sugar_man, bins = [80, 100, 125, 150], rwidth= 0.95) # bin can set the range explicitly
```

```python
plt.hist(blood_sugar_man, bins = [80, 100, 125, 150], rwidth= 0.95, color='g') 
```

```python
blood_sugar_women=[67,98.89,120,133,150,84,69,89,79,120,122,100]
```

```python
plt.hist([blood_sugar_man,blood_sugar_women], bins = [80, 100, 125, 150], rwidth= 0.95, color=['green','orange'],label=['man','women']) 
plt.legend()
plt.xlabel('Sugar range')
plt.ylabel('Total number of patients')
plt.title("Blood Sugar Analysis")
```

```python
plt.hist([blood_sugar_man,blood_sugar_women], bins = [80, 100, 125, 150], rwidth= 0.95, color=['green','orange'],label=['man','women'], orientation="horizontal") 
plt.legend()
plt.xlabel('Sugar range')
plt.ylabel('Total number of patients')
plt.title("Blood Sugar Analysis")
```

<h1>Pie Chart</h1>

```python
expense_value=[1400,600,700,410,750]
expense_labels=["Home Rent","Food","Phone/Internet Bill","Car","Other Utilities"]
```

```python
# sometimes the circle shows up to be slightly oval to fix it use axis equal
plt.pie(expense_value,labels=expense_labels)
plt.axis('equal')
plt.show()
```

```python
plt.pie(expense_value,labels=expense_labels,radius=5)#to enlage the circle use radius
plt.axis('equal')
plt.show()
```

```python
plt.pie(expense_value,labels=expense_labels,radius=5,autopct='%f%%' )#to enlage the circle use radius
plt.axis('equal')
plt.show()
```

```python
plt.pie(expense_value,labels=expense_labels,radius=5,autopct='%0.2f%%' )#to enlage the circle use radius
plt.axis('equal')
plt.show()
```

```python
plt.pie(expense_value,labels=expense_labels,radius=5,autopct='%0.0f%%' )#to enlage the circle use radius
plt.axis('equal')
plt.show()
```

```python
plt.pie(expense_value,labels=expense_labels,radius=5,autopct='%0.0f%%',shadow=True )#to enlage the circle use radius
plt.axis('equal')
plt.show()
```

```python
plt.pie(expense_value,labels=expense_labels,radius=5,autopct='%0.0f%%',shadow=True,explode=[0,0.5,0,0,0] )
plt.axis('equal')
plt.show()
```

```python
plt.pie(expense_value,labels=expense_labels,radius=5,autopct='%0.0f%%',shadow=True,explode=[0,0.5,0,0,0], startangle=180)
plt.axis('equal')
plt.show()
```
